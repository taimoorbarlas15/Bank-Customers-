# Customer segmentation

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Handling missing values
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Fill missing values for numerical columns with mean
for col in numerical_cols:
    df[col].fillna(df[col].mean(), inplace=True)

# Fill missing values for categorical columns with mode
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Standardize numerical columns
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
# Select relevant features for segmentation
features = df[['CreditScore', 'Age', 'EstimatedSalary', 'NumBankAccts', 'NumProducts']]
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Determine the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(features)
    wcss.append(kmeans.inertia_)

# Plot the results of the Elbow Method
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fit the KMeans model with the optimal number of clusters (let's assume 4 for this example)
optimal_clusters = 4
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
df['Cluster'] = kmeans.fit_predict(features)
```

```python
 #Analyze and Visualize the Clusters
```

```python
import seaborn as sns

# Analyze the clusters
cluster_summary = df.groupby('Cluster').mean()

# Visualize the clusters
sns.pairplot(df, hue='Cluster', vars=['CreditScore', 'Age', 'EstimatedSalary', 'NumBankAccts', 'NumProducts'])
plt.show()

# Display cluster summary
cluster_summary
```

# High-Value Customer Identification

```python
# Identifying high-value customers based on transaction patterns and investment behaviors
# Criteria for high-value customers:
# 1. Above average number of transactions
# 2. Above average last transaction amount
# 3. Above average number of bank accounts
# 4. Investments in multiple assets
```

```python
# Define thresholds for high-value criteria
```

```python
avg_num_transactions = transaction_data['NumTransactions'].mean()
avg_last_transaction_amt = transaction_data['LastTransactionAmt'].mean()
avg_num_bank_accts = transaction_data['NumBankAccts'].mean()
investment_threshold = 4  # Considered high-value if invested in 4 or more types
```

```python
# Calculate total investments per customer
```

```python
df['TotalInvestments'] = df[investment_columns].sum(axis=1)
```

```python
# Filter high-value customers
```

```python
high_value_customers = cleaned_data[
    (df['NumTransactions'] > avg_num_transactions) &
    (df['LastTransactionAmt'] > avg_last_transaction_amt) &
    (df['NumBankAccts'] > avg_num_bank_accts) &
    (df['TotalInvestments'] >= investment_threshold)
]
```

```python
# Summary of high-value customers
```

```python
high_value_summary = high_value_customers.describe()
```

```python
print(high_value_summary)
```

```python
# Visualizing some key aspects of high-value customers
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Histogram of Credit Scores for high-value customers
sns.histplot(high_value_customers['CreditScore'], bins=20, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Credit Scores')

# Histogram of Ages for high-value customers
sns.histplot(high_value_customers['Age'], bins=20, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Ages')

# Bar plot of Churn for high-value customers
sns.countplot(x='Churn', data=high_value_customers, ax=axes[1, 1])
axes[1, 1].set_title('Churn Distribution')

plt.tight_layout()
plt.show()
```

High-Value Customer Identification Insights:
Count: There are 779 high-value customers identified based on the criteria set.

Credit Score: They have an average credit score of around 649, similar to the general customer base.

Age: Their average age is approximately 39 years.

Number of Transactions: They engage in a higher number of transactions (average 8.06), compared to the overall average.

Last Transaction Amount: Their last transaction amounts are significantly higher (average $7,640) than the overall average.

Number of Bank Accounts: These customers typically have more bank accounts (average 2.09).
